deque 查找字符

```python
from collections import deque
def S(lines, pattern, n=2):
    pre_lines = deque(maxlen=n)
    for line in lines:
        if pattern in line:
            yield line, pre_lines
            pre_lines.append(line)

with open(r'data/words.txt') as f:
    for line, pre_lines in S(f, 'python', n=2):
        for i in pre_lines:
            print(i,end='')
            print(line, end='')
            print('-' * 20)
        
```

- 实现一个先进先出的dict

```python
from collections import OrderedDict

class O(OrderedDict):
    def __init__(self, maxlen):
        super(O, self).__init__()
        self._maxlen = maxlen
    def __setitem__(self, key, value):
        in_or_notin = 1 if key in self else 0
        if len(self) - in_or_notin >= self._maxlen:
            print('满了，要删掉一个')
            last = self.popitem()
            print('remove:', last)
        if in_or_notin:
            print('删掉原先的然后set操作~')
            del self[key]
            print('set:', (key, value))
        else:
            print('add:', (key, value))
        OrderedDict.__setitem__(self, key, value)
        
```

```python
o = O(3)
o['a'] = 1
o['b'] = 2
o['c'] = 3
o
```

- Iterable

```python
from collections import Iterable
def K(items, ignore=(str, bytes)):
    for x in items:
        if isinstance(x, Iterable) and not isinstance(x, ignore):
            yield from K(x)
        else:
            yield x

items = [1, 2, [3, 4, [5, 6], 7], 8]
for i in K(items):
    print(i)
    
```

- itertools.groupby

```python
from itertools import groupby
str_one = 'AABBCCC'
for k, v in groupby(str_one):
    print("有{}：{}个".format(k, len(list(v))))
```

```python
#假如要忽略大小写
from itertools import groupby
str_two = 'aAbBCCC'
for k, v in groupby(str_two, lambda x: x.upper()):
    print("有{}：{}个".format(k, len(list(v))))
```

```python
from itertools import groupby

rows = [
    {'address': '5412 N CLARK', 'date': '07/01/2012'},
    {'address': '5148 N CLARK', 'date': '07/04/2012'},
    {'address': '5800 E 58TH', 'date': '07/02/2012'},
    {'address': '2122 N CLARK', 'date': '07/03/2012'},
    {'address': '5645 N RAVENSWOOD', 'date': '07/02/2012'},
    {'address': '1060 W ADDISON', 'date': '07/02/2012'},
    {'address': '4801 N BROADWAY', 'date': '07/01/2012'},
    {'address': '1039 W GRANVILLE', 'date': '07/04/2012'},
]

rows.sort(key=lambda x: x['date'])
for date, items in groupby(rows, lambda x: x['date']):
    print(date)
    for i in items:
        print(' -', i)
```
