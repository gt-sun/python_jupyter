deque 查找字符

```python
from collections import deque
def S(lines, pattern, n=2):
    pre_lines = deque(maxlen=n)
    for line in lines:
        if pattern in line:
            yield line, pre_lines
            pre_lines.append(line)

with open(r'data/words.txt') as f:
    for line, pre_lines in S(f, 'python', n=2):
        for i in pre_lines:
            print(i,end='')
            print(line, end='')
            print('-' * 20)
        
```
