```python
def fun():
    l_one = [31, -41, 59, 26, -53, 58, 97, -93, -23, 84] * 100
    maxsofar = 0
    n = len(l_one)
    for i in range(0, n):
        for j in range(i, n):
            sum = 0
            for k in range(i, j+1):
                sum += l_one[k]
            maxsofar = max(maxsofar, sum)
    return maxsofar
```

法4： 扫描算法

线性算法

```python
def fun():
    items = [31, -41, 59, 26, -53, 58, 97, -93, -23, 84] * 100
    n = len(items)
    maxsofar = 0
    maxendinghere = 0
    for i in range(0, n):
        maxendinghere = max(maxendinghere + items[i], 0)
        maxsofar = max(maxendinghere, maxsofar)
    return maxsofar
```

```python
%time fun()
```

```python
def foo(x, t):
    n = len(x)
    l = -1
    u = n
    while l + 1 != u:
        m = (l + u) // 2
        if x[m] < t:
            l = m
        else:
            u = m
    p = u
    if p >= n or x[p] != t:
        p = -1 #不存在
    print(p)
```
